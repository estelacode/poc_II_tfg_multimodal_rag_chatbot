### Gradio Multimodal Chatbot
Display text and media files (audio, video, images)

`poetry add gradio_multimodalchatbot`

```python
import gradio as gr
from gradio_multimodalchatbot import MultimodalChatbot
from gradio.data_classes import FileData
```

```python
def rag_respond(message, history):
    return 'Hola'
```

```python

demo = gr.ChatInterface(
    fn=rag_respond, 
    type="messages", 
    examples=[{"text": "Hello", "files": []}], 
    title="Multimodal Chatbot", multimodal=True)
```

```python
demo.launch()
```
